## QQ Plot 
- The QQ plots require me to remove all values of -1.0 from the distribution as -1.0 was reported in the snptest code where the software could not fit a model so output values were replaced with -1.0
- Then use a git hub piece of code: https://github.com/khramts/assocplots/blob/master/Tutorial.ipynb
- this means I can plot the QQ plot and Manhatten plots 

```python
remove_df = df[df['STAT_SCORE_frequentist_add_center_pc1_pc2_pc3_pc4_pc5_pc6_pc7_pc8_pc9_pc10_expected_pvalue'] != -1.00000].copy()
```

```python
remove_df
```

```python
# This is an example of a classic QQ plot with 95% confidence interval plotted for the null distribution
from assocplots.qqplot import *
import matplotlib as mpl
mpl.rcParams['figure.dpi']=100
mpl.rcParams['savefig.dpi']=100
mpl.rcParams['figure.figsize']= 3.8, 3.8

qqplot([remove_df['STAT_SCORE_frequentist_add_center_pc1_pc2_pc3_pc4_pc5_pc6_pc7_pc8_pc9_pc10_expected_pvalue']], 
       ['STATacute'], 
       color=['b'], 
       fill_dens=[0.2], 
       error_type='theoretical', 
       distribution='beta',
       title='STATacute ')
```

```python
# Now we want to calculate the genomic control (inflation factor, lambda)
get_lambda(remove_df['STAT_SCORE_frequentist_add_center_pc1_pc2_pc3_pc4_pc5_pc6_pc7_pc8_pc9_pc10_expected_pvalue'], definition = 'median')
```
